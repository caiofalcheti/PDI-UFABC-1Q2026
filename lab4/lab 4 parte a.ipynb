```python


import cv2
import numpy as np
import sys

# tamanhos de kernel desejados
kernels = [3, 5, 11, 29]

# carregar imagem
filename = 'foto_grupo_2.png'
src = cv2.imread(filename)

if src is None:
    print("Erro ao abrir a imagem")
    sys.exit(1)

for k in kernels:

    # ----- Filtro de média -----
    kernel = np.ones((k, k), np.float32) / (k * k)
    media = cv2.filter2D(src, -1, kernel)
    cv2.imwrite(f"media_{k}x{k}.jpg", media)

    # ----- Filtro Gaussiano -----
    gaussian = cv2.GaussianBlur(src, (k, k), 0)
    cv2.imwrite(f"gaussian_{k}x{k}.jpg", gaussian)

    # ----- Filtro Mediana -----
    median = cv2.medianBlur(src, k)
    cv2.imwrite(f"median_{k}x{k}.jpg", median)

    # ----- Filtro Bilateral -----
    bilateral = cv2.bilateralFilter(src, k, k*2, k//2)
    cv2.imwrite(f"bilateral_{k}x{k}.jpg", bilateral)

print("Processamento concluído. Imagens salvas.")
```
