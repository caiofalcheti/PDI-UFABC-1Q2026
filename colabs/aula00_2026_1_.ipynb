# Antes de começar: Notebooks em Python

O que é este documento? Você sabe o que é? Já ouviu falar em _**[Literate Programming](http://cursos.leg.ufpr.br/prr/capLitProg.html)**_? (Programação literária, alfabética, culta?)

Isto se chama _notebook_: contém textos explicativos (**células de texto** no formato [Markdown](https://colab.research.google.com/notebooks/markdown_guide.ipynb)) e códigos de programas em [Python](https://www.python.org/) intercalados (**células de código**).

É fácil distinguir as células de código: elas vem precedidas por

    [ ]
    
como a célula abaixo. Clique nela.

```python
2 + 2
```

Para executar uma célula de código, basta selecionar a célula e digitar <kbd>SHIFT+ENTER</kbd>, mas antes:

- Se você estiver no Colab, você também pode executar uma célula clicando no botão de *play* que aparece quando você coloca o mouse sobre os colchetes ``[ ]``

- Se você estiver no Colab no **celular**, o <kbd>SHIFT+ENTER</kbd> pode não funcionar. Nesse caso, execute apertando o *play*.

- Você também pode rodar um arquivo Notebook (ex. arquivo **aula01.ipynb**) localmente, instalando o [jupyter notebook](https://jupyter.org).

- Para rodar o arquivo **aula01.ipynb** no Colab do seu navegador, salvar uma cópia em Arquivo -> Salvar uma cópia no Drive; Abrir o arquivos **Cópia de aula01.ipynb**, localizado na pasta **Colab Notebooks** no seu Drive.


O resultado será exibido logo após a execução. Execute a célula acima e veja o resultado.


Neste _notebook_, vamos aprender alguns elementos básicos de programação em Python e vamos nos familiarizar com o ambiente.
